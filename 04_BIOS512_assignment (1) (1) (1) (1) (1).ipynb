```r
library('tidyverse')
```

```r
wind = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv', col_types = cols())
```

```r
wind %>% head
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_freq = wind %>% 

    group_by(province_territory) %>% 
    summarize(frequency = n()) %>%
    arrange(frequency) %>%
    mutate(province_territory = province_territory %>% 
    fct_inorder())

wind_freq
```

```r
p = ggplot(wind_freq, aes (y = province_territory, x = frequency)) +
geom_col()

p
```

Make a bar **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind_freq2 = wind %>% 

    mutate(manufacturer = manufacturer %>% fct_lump(n = 5)) %>% 
    group_by(manufacturer) %>% 
    summarize(frequency = n()) %>%
    arrange(frequency) %>%
    mutate(manufacturer = manufacturer %>% fct_inorder() %>% fct_rev()) %>%
    mutate(manufacturer = manufacturer %>% fct_relevel('Other', after= Inf)) 

wind_freq2
```

```r
p = ggplot(wind_freq2, aes (y = manufacturer, x = frequency)) +
geom_col()

p
```
